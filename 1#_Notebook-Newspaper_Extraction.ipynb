The next step is to get the summaries themselves, by using the Newspaper Library.

```python
def get_textbody(row):
    try:
        url = row['url']
        article = Article(url)
        article.download()
        article.parse()
        article.nlp()
#        summary = article.summary
        text = article.text        
        return text
    
    except :
        pass

newspaperdata['Text'] = newspaperdata.apply(get_textbody, axis=1)
```

Finishing up this part of the extraction, I'm going to drop all rows that have not retrieved a Summary and use the shape again to know how many I'm left with.

```python
newspaperdata = newspaperdata.dropna()
print(newspaperdata.shape)
newspaperdata
```

Finally, I'm using NLTK to get the summaries normalized. After they are normalized, I can use the drop to remove the column "Summary".

```python
def normalized_text(row):
    content = row['Text']
    lowered = content.lower()
    tokens = nltk.tokenize.word_tokenize(lowered)
    keywords = [keyword for keyword in tokens if keyword.isalpha()]
    keywords_string = ' '.join(keywords)
    return keywords_string

#def normalized_title(row):
#    content = row['title']
#    lowered = content.lower()
#    tokens = nltk.tokenize.word_tokenize(lowered)
#    keywords = [keyword for keyword in tokens if keyword.isalpha()]
#    keywords_string = ' '.join(keywords)
#    return keywords_string
```

```python
newspaperdata['NormalizedText'] = newspaperdata.apply(normalized_text, axis=1)
#newspaperdata['NormalizedTitle'] = newspaperdata.apply(normalized_title, axis=1)
```
